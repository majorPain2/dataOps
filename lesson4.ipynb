```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def doc_check(docnum):
#    docnum = input('Введи номер документа: ')
    for list_ in documents:
        if list_['number'] == docnum:
            return list_
    return {'name':'Нет такого документа'}

def dir_check(docnum):
  #  docnum = input('Введи номер документа: ')
    for shelf, doc in directories.items():
        if docnum in doc:
            return shelf
    return 'Нет такого документа'

def list_doc():
    for list_ in documents:
        docnum = list_["number"]
        for shelf, doc in directories.items():
            if docnum in doc:
                print(f'№: {list_["number"]}, тип: {list_["type"]}, владелец:{list_["name"]}, полка: {shelf}')
 
def del_shelf(shelfnum):
    for shelf, doc in directories.items():
        if int(shelf) == shelfnum:
            if len(doc) > 0:
                return 0
            else:
                return shelfnum
    return -1
              
def add_shelf():
    new_shelf = input('Введи номер полки: ')
    if new_shelf in directories:
        print('Такая полка есть')
    else:
        directories[new_shelf] = []
        print('Создали новую полку')
    return new_shelf
              
def add_doc(docnum):
    doc_type = input('Введи тип документа: ')
    doc_name = input('Введи имя: ')
    documents.append({'type': doc_type, 'number': docnum, 'name': doc_name})
```

```python
def main():
    while True:
        command = input('Введи команду: ')
        if command == 'q':
            break
        elif command == 'p':
            docnum = input('Введи номер документа: ')
            print(doc_check(docnum)['name'])
        elif command == 's':
            docnum = input('Введи номер документа: ')
            print(dir_check(docnum))
        elif command == 'l':
            list_doc()
        elif command == 'ads':
            add_shelf()
        elif command == 'ds':
            shelfnum = int(input('Введи номер полки: '))
            a = del_shelf(shelfnum)
            if a == 0:
                print('На полке есть документы, удалите их перед удалением полки.')
            elif a == -1:
                print('Такой полки не существует.')
            else:
                del directories[str(a)]
                print('Полка удалена.')
        elif command == 'ad':   
            docnum = input('Введи номер документа: ')
            if doc_check(docnum)['name'] == 'Нет такого документа':
                add_doc(docnum)
                shelf = add_shelf()
                directories[shelf].append(docnum) 
            else:
                print('Такой документ уже существует')
        elif command == 'd':
            docnum = input('Введи номер документа: ')
            try:
                documents.remove(doc_check(docnum))
                a = dir_check(docnum)
                directories[a].remove(docnum)
            except:
                print('Нет такого документа')
        elif command == 'm':       
            docnum = input('Введи номер документа: ')
            try:
                a = dir_check(docnum)
                directories[a].remove(docnum)
                shelf = add_shelf()
                directories[shelf].append(docnum)
            except:
                print('Нет такого документа')        
        else: print('Нет такой команды')
            
main()
```
